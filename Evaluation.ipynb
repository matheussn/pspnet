```python
import matplotlib.pyplot as plt
import json
```

```python
file = open('None.log.json')
```

```python
lines = file.readlines()
```

```python
m_dice = []
m_acc = []
loss = []

for line in lines:
    json_line = json.loads(line)
    if len(json_line):
        if json_line['mode'] == 'train':
            loss.append(json_line['loss'])
        else:
            m_acc.append(json_line['mAcc'])
            m_dice.append(json_line['mDice'])
```

```python
len(loss)
```

```python
len(m_acc)
```

```python
range(1, len(m_acc))
```

```python
plt.figure(figsize=(15, 5))
plt.plot(range(1, len(m_acc)+1), loss, label="loss")
plt.plot(range(1, len(m_acc)+1), m_acc, label="M Acc")

plt.xticks(range(1, len(m_acc)+1), rotation=90)

plt.title("Training Loss")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.show()
# plt.savefig(f'{base_dir}/losses_100.png')
plt.close()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(range(1, len(m_acc)+1), m_dice, label="M Dice")
plt.xticks(range(1, len(m_acc)+1), rotation=90)
plt.title("Training Loss")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.show()
plt.close()
```
